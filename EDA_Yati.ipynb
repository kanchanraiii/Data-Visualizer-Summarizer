```python
rating_counts = df['survey_response_rating'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(rating_counts, labels = rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.legend(title="Ratings", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Survey Response Ratings')
plt.show()
```

```python
# Filter data -- only those who have joined the event
joined_event = df[df['has_joined_event'] == 'Yes']

rating_count = joined_event['survey_response_rating'].notna().sum()
feedback_count = joined_event['survey_response_feedback'].notna().sum()
```

```python
print(rating_count)
print(feedback_count)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=['Rating', 'Feedback'], y=[rating_count, feedback_count])
plt.title('Count of students sho joined the event and provided survey responses')
plt.xlabel('Response Type')
plt.ylabel('Count')
plt.show()
```

```python
rating_mean = joined_event['survey_response_rating'].mean()
rating_std = joined_event['survey_response_rating'].std()

print("Survey Response Rating Count:",rating_count)
print("Survey Response Feedback Count: ",feedback_count)
print("Survey Response Rating Mean:",rating_mean)
```

```python
def standardize_designation(designation):
    if pd.isna(designation):
        return 'Unknown'
    designation = designation.strip().lower()
    if 'student' in designation:
        return 'Student'
    elif 'ar' in designation or 'vr' in designation or 'xr' in designation :
        return 'AR/VR Developer'
    elif 'sde' in designation or 'developer' in designation or 'software engineer' in designation :
        return 'Software Developer'
    elif 'ui' in designation or 'ux' in designation :
        return 'UI/UX Designer'
    elif 'founder' in designation:
        return 'Founder'
    else:
        return 'Other'

df['Standardized Designation'] = df['What is your designation?'].apply(standardize_designation)
```

```python
df['Standardized Designation'].unique()
```

```python
# Plotting the distribution of designations
plt.figure(figsize=(30, 6))
sns.countplot(data=df, x='What is your designation?')
plt.xlabel('Designation')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```
