#### Counting the different types of designations people have filled in the form

```python
designation_counts=df["What is your designation?"].value_counts()
designation_counts
```

#### Normalizing the designations column

```python
designation_dict = designation_counts.to_dict()
designation_dict
```

```python
total_count = sum(designation_dict.values())
normalized_dict = {k: (v / total_count) * 100 for k, v in designation_dict.items()}
normalized_dict
```

### Normalized Distribution of Designations

```python
plt.figure(figsize=(12, 8))
plt.bar(normalized_dict.keys(), normalized_dict.values(),edgecolor='black')
plt.title('Normalized Distribution of Designations')
plt.xlabel('Designattion')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()
```

### Top Designations

```python
designation_counts = pd.Series(designation_dict)

# Threshold for grouping less frequent categories into 'Others'
threshold = 1.0  # Adjust as needed based on your dataset

# Combine less frequent categories into 'Others'
designation_counts_others = designation_counts[designation_counts <= threshold].sum()
top_designations = designation_counts[designation_counts > threshold]
top_designations['Others'] = designation_counts_others

# Sort the values
top_designations = top_designations.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(8, 6))
top_designations.plot(kind='barh', color='skyblue', edgecolor='black')

# Add titles and labels
plt.title('Top Designations')
plt.xlabel('Percentage (%)')
plt.ylabel('Designation')

# Show the plot
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist

# Download NLTK resources (only needed once)
nltk.download('punkt')
nltk.download('stopwords')


survey_feedback = df["survey_response_feedback"].dropna().astype(str)

# Function to tokenize text into sentences and summarize
def summarize_feedback(text):
    try:
        sentences = sent_tokenize(text)
        words = nltk.word_tokenize(text)
        words = [word.lower() for word in words if word.isalnum()]
        stop_words = set(stopwords.words('english'))
        words = [word for word in words if word not in stop_words]
        freqdist = FreqDist(words)
        summary = freqdist.most_common(1)
        return summary
    except Exception as e:
        print(f"Error: {e}")
        return None

# Apply summarization function to survey_feedback Series
summarized_results = survey_feedback.apply(summarize_feedback)

# Create a new DataFrame to store the original feedback and summaries
summary_df = pd.DataFrame({
    'Original Feedback': survey_feedback,
    'Summary': summarized_results
})

# Display the summarized feedback
print(summary_df)
```
